Вы работаете секретарем и вам постоянно приходят различные документы. Вы должны быть очень внимательны, чтобы не потерять ни один документ. Каталог документов хранится в следующем виде:

```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
```

Перечень полок, на которых находятся документы хранится в следующем виде:

```python
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
```

Необходимо реализовать пользовательские команды (внимание! это не название функций, которые должны быть выразительными, а команды, которые вводит пользователь, чтобы получить необходимый результат):
  
- p – people – команда, которая спросит номер документа и выведет имя человека, которому он принадлежит;  
- s – shelf – команда, которая спросит номер документа и выведет номер полки, на которой он находится  
Правильно обработайте ситуацию, когда пользователь будет вводить несуществующий документ.
- l – list – команда, которая выведет список всех документов в формате   
**passport "2207 876234" "Василий Гупкин"**  
- as – add shelf – команда, которая спросит номер новой полки и добавит ее в перечень
- d – delete – команда, которая спросит номер документа и удалит его из каталога и из перечня полок. Предусмотрите сценарий, когда пользователь вводит несуществующий документ;  
- m – move – команда, которая спросит номер документа и целевую полку и переместит его с текущей полки на целевую. Корректно обработайте кейсы, когда пользователь пытается переместить несуществующий документ или переместить документ на несуществующую полку;  
- a – add – команда, которая добавит новый документ в каталог и в перечень полок, спросив его номер, тип, имя владельца и номер полки, на котором он будет храниться.

```python
# p – people – команда, которая спросит номер документа и выведет имя человека, которому он принадлежит; 
def people_name(list_doc):
    doc_num = input("Введите номер документа: ")
    for x in list_doc:
        if doc_num == x['number']:
            return print(x['name'])
    return print('Такого номера документа нет в базе!')

# s – shelf – команда, которая спросит номер документа и выведет номер полки, на которой он находится
def shelf_num(dict_dir):
    doc_num = input("Введите номер документа: ")
    for key, value in dict_dir.items():
        for num in value:
            if num == doc_num:
                return print(key)
    return print('Такого номера документа нет в базе!')

# l – list – команда, которая выведет список всех документов в формате
def lists(list_doc):
    for x in list_doc:
        y = list(x.values())
        print(f'{y[0]} "{y[1]}" "{y[2]}')
    return

#as – add shelf – команда, которая спросит номер новой полки и добавит ее в перечень    
def add_shelf(dict_dir):
    shelf_num = input("Введите номер новой полки: ")
    if shelf_num in dict_dir.keys():
        print("Такая полка уже существует!")
    else:
        dict_dir[shelf_num] = []

# d – delete – команда, которая спросит номер документа и удалит его из каталога и из перечня полок. 
def del_doc(list_doc, dict_dir):
    doc_num = input("Введите номер документа для удаления из базы: ")
    for x in list_doc:
        if doc_num == x['number']:
            list_doc.remove(x)
            for v in dict_dir.values():
                if doc_num in v:
                    v.remove(doc_num)
                    return
    return print('Такого номера документа нет в базе!')

#m – move – команда, которая спросит номер документа и целевую полку и переместит его с текущей полки на целевую. 
def move_doc(dict_dir):
    doc_num = input("Введите номер документа: ")
    shelf_num = input("Введите номер полки, на которую надо переместить указанный документ: ")
    if len([v for v in dict_dir.values() if doc_num in v]) == 0:
        return print("Такого доумента не существует!")
    else: 
        if shelf_num not in dict_dir.keys():
            return print("Такой полки не существует!")
        else:
            [v for v in dict_dir.values() if doc_num in v][0].remove(doc_num)
            dict_dir[shelf_num].append(doc_num)
    return print("Документ перемещён успешно!")

# a – add – команда, которая добавит новый документ в каталог и в перечень полок, 
# спросив его номер, тип, имя владельца и номер полки, на котором он будет храниться.        

# def doc_scan(dict_dir): # Вспомогательная функция1
#     doc_num = input("Введите номер документа: ")
#     if len([v for v in dict_dir.values() if doc_num in v]) > 0:
#         YesNo = input('Такой номер документа уже есть в базе! Заменить данные? (y/n)') 

# def add_doc_func(list_doc, dict_dir): # Вспомогательная функция2
#             doc_type = input("Введите тип документа: ")
#             name = input("Введите имя и фамилию владельца документа: ")
#             shelf_num = input("Введите номер полки: ")
# # Добавление документа в список документов
#             if len([v for v in dict_dir.values() if doc_num in v]) > 0:
#                 list_doc.remove([d for d in list_doc if d['number'] == doc_num][0])
#                 [v for v in dict_dir.values() if doc_num in v][0].remove(doc_num)
#             new_dict = {}
#             new_dict['type'] = doc_type
#             new_dict['number'] = doc_num
#             new_dict['name'] = name
#             list_doc.append(new_dict)
# # Добавление документа на полку   
#             if shelf_num in dict_dir.keys():
#                 dict_dir[shelf_num].append(doc_num) 
#             else:
#                 dict_dir[shelf_num]=[doc_num]
#             return print('Новые данные успешно добавлены в базу!')
        
def add_doc(list_doc, dict_dir):
# Пользовательский ввод
    doc_num = input("Введите номер документа: ")
    if len([v for v in dict_dir.values() if doc_num in v]) > 0:
        YesNo = input('Такой номер документа уже есть в базе! Заменить данные? (y/n)') 
        if YesNo == 'n':
            return
        else:
            doc_type = input("Введите тип документа: ")
            name = input("Введите имя и фамилию владельца документа: ")
            shelf_num = input("Введите номер полки: ")
# Добавление документа в список документов
            if len([v for v in dict_dir.values() if doc_num in v]) > 0:
                list_doc.remove([d for d in list_doc if d['number'] == doc_num][0])
                [v for v in dict_dir.values() if doc_num in v][0].remove(doc_num)
            new_dict = {}
            new_dict['type'] = doc_type
            new_dict['number'] = doc_num
            new_dict['name'] = name
            list_doc.append(new_dict)
# Добавление документа на полку   
            if shelf_num in dict_dir.keys():
                dict_dir[shelf_num].append(doc_num) 
            else:
                dict_dir[shelf_num]=[doc_num]
            return print('Новые данные успешно добавлены в базу!')
    else:
        doc_type = input("Введите тип документа: ")
        name = input("Введите имя и фамилию владельца документа: ")
        shelf_num = input("Введите номер полки: ")
# Добавление документа в список документов
        if len([v for v in dict_dir.values() if doc_num in v]) > 0:
            list_doc.remove([d for d in list_doc if d['number'] == doc_num][0])
            [v for v in dict_dir.values() if doc_num in v][0].remove(doc_num)
        new_dict = {}
        new_dict['type'] = doc_type
        new_dict['number'] = doc_num
        new_dict['name'] = name
        list_doc.append(new_dict)
# Добавление документа на полку   
        if shelf_num in dict_dir.keys():
            dict_dir[shelf_num].append(doc_num) 
        else:
            dict_dir[shelf_num]=[doc_num]
        return print('Новые данные успешно добавлены в базу!')
  

#people_name(documents)
#shelf_num(directories)
#lists(documents)
#add_shelf(directories)
#del_doc(documents,directories)
#move_doc(directories)
#add_doc(documents,directories)
#print(documents,directories,sep='\n')
```

```python
def main():
    print("""
    Пользовательские Команды для работы с базой данных:
    p – команда, которая спросит номер документа и выведет имя человека, которому он принадлежит;
    s – команда, которая спросит номер документа и выведет номер полки, на которой он находится;
    l – команда, которая выведет список всех документов; 
    as – команда, которая спросит номер новой полки и добавит ее в перечень;
    d – команда, которая спросит номер документа и удалит его из каталога и из перечня полок;
    m – команда, которая спросит номер документа и целевую полку и переместит его с текущей полки на целевую;
    a – команда, которая добавит новый документ в каталог и в перечень полок, спросив его номер, тип, имя владельца и номер полки, на котором он будет храниться.
    q - выход из системы;
    """)
    while True:
        user_input = input('Введите команду из перечня команд: ') 
        if user_input == 'p':
            people_name(documents)
        elif user_input == 's':
            shelf_num(directories)
        elif user_input == 'l':
            lists(documents)
        elif user_input == 'as':  
            add_shelf(directories)
        elif user_input == 'd': 
            del_doc(documents,directories)
        elif user_input == 'm': 
            move_doc(directories)
        elif user_input == 'a':
            add_doc(documents,directories)
        elif user_input == 'q':
            print('До свидания!')
            break

main()
```
